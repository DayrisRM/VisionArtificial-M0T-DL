```python
import cv2
import numpy as np
import copy
from matplotlib import pyplot as plt
```

```python
def pintarSegmento(img, pt1, pt2, color):
  #cv2.line(img, tuple(pt1), tuple(pt2), (0, 0, 255), 2)
  cv2.line(img, pt1, pt2, color, 2)
  cv2.circle(img, pt1, 3, color ,cv2.FILLED)
  cv2.circle(img, pt2, 3, color ,cv2.FILLED)
```

```python
def pintarTrazo(img, trazo, color):
  if len(trazo) == 0:
    return()
  else:
    print(trazo)
  if len(trazo) == 1:
    #cv2.circle(img, (50, 50), 5, (0,0,255) ,cv2.FILLED)
    cv2.circle(img, trazo[0], 5, color ,cv2.FILLED)
  else:
    for it in range(1, len(trazo)):
      pintarSegmento(img, tuple(trazo[it-1]),tuple(trazo[it]), (0, 0, 255))
color = [0,0,255]
p = (29,445)
```
